### Ejercicio 1

Escriba una función que tome una serie y devuelva el 10% superior de los registros. En este caso, los asalariados.

```python
salaries = pd.Series([150000,90000,120000,30000,10000,5000,40000,50000, 80000,35000,27000,14000,28000,22000,25000])
```

````python
salaries = pd.Series([150000,90000,120000,30000,10000,5000,40000,50000,80000,35000,27000,14000,28000,22000,25000])
salaries.head()
````

````python
valor = 0
def diez_sup(m):
    valor = int(len(m) / 10)
    return m.sort_values(ascending = False).head(valor)
````

````python
diez_sup(salaries)
````
