# Libreria Matplotlib

Matplotlib, es una biblioteca de python que se utiliza para crear gráficos y diagramas 2D mediante el uso de scripts de python. Tiene un módulo llamado pyplot que facilita las tareas de trazado al proporcionar una función para controlar estilos de línea, propiedades de fuente, ejes de formato, etc. Admite una gran variedad de gráficos y diagramas, como histograma, gráficos de barras, espectros de potencia, gráficos de errores, etc. Se usa junto con NumPy para proporcionar un entorno que sea una alternativa de código abierto eficaz para MatLab. También se puede utilizar con kits de herramientas de gráficos como PyQt y wxPython.

## Trabajando con Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

import numpy as np
```

### `Lines`

```python
x = np.linspace(0,10,50)
y = np.sin(x)

plt.plot(x,y,color = 'green')
plt.legend('SIN',loc = 2)
```

### `Scatterplots`

Aquí estamos usando un ruido aleatorio ("jitter") agregado sobre $f(x) = 2x + 20$. La Serie y2 para obtener datos más ruidosos y de apariencia más real.

```python
x2 = np.random.randn(100) * 20 + 50
y2 = 2 * x2 + 20

jitter = np.random.randn(100) * 20

y2_randomize = y2 + jitter
plt.scatter(x2,y2_randomize)
```

### `Histogram`

Ejemplo con histograma utilizando los datos x2 (una distribución aleatoria)

```python
plt.hist(x2)
```

Ejemplo 2

```python
from scipy.stats import norm
```

```python
#Gaussian distribution parameters
nbins=30
mu=50
sigma=20
dist= norm(mu,sigma)
N = 1000

xnorm = np.linspace(-20,120,nbins)
x3 = np.random.randn(N) * 20 + 50
ynorm = dist.pdf(xnorm)

plt.hist(x3,bins=nbins,normed=1,label='Histograma');
plt.plot(xnorm,ynorm,c = 'black',label = r'$\mu=%i,\ \sigma=%.1f$' % (mu,sigma))
plt.legend()
```

### `Subplot()`

Podemos agregarlos con `figure.add_subplot()` o crear una figura con sus subplots directamente llamando a `plt.subplots()`

```python
figure, axeses = plt.subplots(2,2)  # or axs = plt.subplots(2,2)
```

```python
figure, axs = plt.subplots(2,2)

axs[0,0].plot(x,y)
axs[0,1].scatter(x2,y2_randomize)
axs[0,1].plot(x2,y2,c='blue')
axs[1,0].plot(x,x ** 2)
axs[1,1].plot(x2,x2)

figure.set_size_inches(10,6)
axs[0,1].legend(['lineal function'])
```

### Fill_Between

Podemos rellenar áreas de la parcela con color sólido.
La sintaxis es `plt.fill_between(x,y1,y2)`
Donde y1 e y2 son valores `np.array` o escalares.

```python
plt.fill_between(x,y,y/2)
```

### fill_between(where=)

Podemos rellenar áreas de la trama con colores sólidos de manera diferente en diferentes partes de la gráfica.

La sintaxis es `plt.fill_between (X, Y1, Y2, donde = boolean_array)`

Donde y1 e y2 son valores `np.array` o escalares, y boolean_array se puede construir a partir de x, de y, o de lo que quieras.

### Marcadores y estilos de línea.

```python
figure, axes = plt.subplots(1,1)

x = np.arange(0,3,.2)
y1 = x
y2 = x ** 2
y3 = x ** 3

axes.plot(x,y1,'r--')
axes.plot(x,y2,'bs')
axes.plot(x,y3,'g^')

plt.show()
```

### Legends

```python
x = np.linspace(-np.pi,np.pi,100)
c,s = np.cos(x),np.sin(x)

plt.plot(x,c,label='cosine')
plt.plot(x,s,label='sine')

plt.legend(loc='best',frameon=False)
```

### Ticks, tick labels, limits

### Ticks

```python
figure, axes = plt.subplots(1,1)

x = np.arange(0,3,.2)
y = x ** 3

axes.plot(x,y)
axes.set_xticks(range(4))
```

### Tick Label

```python
figure, axes = plt.subplots(1,1)

x = np.arange(0, 3, .2)
y = x ** 3

axes.plot(x,y)
axes.set_xticks(range(4))
axes.set_xticklabels(list('ABCD'))
```

### Limits

```python
plt.plot(x,y)

axes = plt.gca()
axes.set_xlim((0.5,2))
axes.set_ylim((0,10))
```
